```python
def fix_dataset_format(hf_dataset):
    """Fix dataset format to ensure prompt and completion are lists."""
    # Ensure prompt and completion are in lists format
    if isinstance(hf_dataset["prompt"], dict):
        hf_dataset["prompt"] = [hf_dataset["prompt"]]
    if isinstance(hf_dataset["completion"], dict):
        hf_dataset["completion"] = [hf_dataset["completion"]]
    return hf_dataset

train_dataset = ds.map(fix_dataset_format)
```
